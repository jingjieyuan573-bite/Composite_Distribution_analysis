```python
import yfinance as yf
import pandas as pd

# ==========================================================
# 指数列表（原始收盘价）
# ==========================================================
tickers = {
    "SP500": "^GSPC",
    "HSI": "^HSI",
    "NG": "NG=F",
    "EURUSD": "EURUSD=X"
}

start_date = "2020-01-01"
end_date = "2025-06-30"

# ==========================================================
# 下载数据
# ==========================================================
all_close = pd.DataFrame()

for name, ticker in tickers.items():
    print(f"Downloading {name} ({ticker})...")
    df = yf.download(ticker, start=start_date, end=end_date)
    if "Close" not in df.columns:
        raise ValueError(f"{ticker} 没有收盘价列")
    all_close[name] = df["Close"]

# 对齐交易日（去掉缺失）
all_close.dropna(inplace=True)

print("\n收盘价前5行：")
print(all_close.head())

# 保存原始收盘价到 CSV
all_close.to_csv("four_close.csv")
print("\n原始收盘价已保存到 four_close.csv")
```
